```python
import pandas as pd
df=pd.read_csv('csv_데이터중복성체크.csv',index_col=0)
df
```

```python
df.info()
```

```python
# 7월31일의 값이 2개임. 즉 중복된 인덱스값이 있음.
# 중복된 index값이 있으면 데이터분석(corr:상관계수)등이 안될수 있음.
df.index.value_counts()   
```

```python
df.loc['07월 31일']
```

```python
df=df.drop_duplicates() # 메서드는 내용이 중복되는 행을 제거하는 메서드입니다
df
```

```python
#df.drop_duplicates(inplace=True)
df.index.value_counts() # 중복행갯수확인
```

```python
# 행단위의 중복은 제거했음
# 만약 열단위의 중복을 확인하고자 한다면 어떻게 하면 될까요?
```

```python
df=df.T.drop_duplicates().T
df
```

```python
df.corr()
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

plt.figure(figsize=(15,15))
sns.heatmap(data = df.corr(), annot=True,fmt = '.2f', linewidths=.5, cmap='Blues')
```
