```python
import folium

# 지도의 중심을 설정합니다.
latitude, longitude = 37.5665, 126.9780  # 서울의 위도, 경도

# Folium을 사용하여 지도를 만듭니다.
map_osm = folium.Map(location=[latitude, longitude], 
                     zoom_start=70,
                    width=300,height=200)

map_osm
```

```python
import pandas as pd
df=pd.DataFrame({'위도':[37.50140089999999,
                         37.644936799999996, 
                         37.6403353, 
                         37.645723499999995,
                         37.6597831],
                 
                  '경도':[127.1586471,
                          127.0737283, 
                          127.0776372,
                          127.02044939999999,
                          127.0311865],
                '장소명':['장소1','장소2','장소3','장소4','장소5'],
                '비고':['비고1','비고2','비고3','비고4','비고5']})
df
```

```python
## 기본 차트 출력
lines=df[['위도', '경도']]
center = [37.541, 126.986]
m = folium.Map(location=center, zoom_start=10) 
m
```

```python
# 원차트
m = folium.Map(location=center, zoom_start=10) 

folium.CircleMarker(
    location = df.loc[0, ['위도', '경도']],
    radius = 100,
     weight = 10, # 선두꼐
).add_to(m)

m
```

```python
# 패스차트
# 아래의 m을 생성하지 않으면 원위에 패스가 추가됨
m = folium.Map(location=center, zoom_start=10) 

folium.PolyLine(
    locations = lines,
    tooltip = 'PolyLine'
).add_to(m)

m
```

```python
# 다각형 폴리 차트
lines=df[['위도', '경도']]
center = [37.541, 126.986]
m = folium.Map(location=center, zoom_start=10) 

folium.Polygon(
    locations = lines,
    fill = True,
    tooltip = 'Polygon'
).add_to(m)

m
```

```python
# 위경도 위치에 마커출력
# 마우스를 올려놓으면 장소명이 출력되고 (tooltip)
# 마우스를 클릭하면 비고가 출력됨(popup)

center = [37.541, 126.986]
m = folium.Map(location=center, zoom_start=10,width=300,height=200)

# 1000 개의 데이터만 그려냅니다.
for i in df.index[:]:
    folium.Marker(
        location = df.loc[i, ['위도', '경도']],     
        tooltip = df.loc[i, '장소명'],
        popup=df.loc[i, '비고']
    ).add_to(m)
m
```

```python
# 위경도 위치에 마커출력
# 마우스를 올려놓으면 장소명이 출력되고 (tooltip)
# 마우스를 클릭하면 비고가 출력됨(popup)

center = [37.541, 126.986]
m = folium.Map(location=center, zoom_start=10)
folium.TileLayer("Stamen Terrain").add_to(m)

# 1000 개의 데이터만 그려냅니다.
for i in df.index[:]:
    folium.Marker(
        location = df.loc[i, ['위도', '경도']],     
        tooltip = df.loc[i, '장소명'],
        popup=df.loc[i, '비고']
    ).add_to(m)



m
```
